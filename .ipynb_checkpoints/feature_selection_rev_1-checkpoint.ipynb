```python
### Basic Setup

import warnings

with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    import numpy as np
    
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

pd.options.display.max_columns = 0
```

```python
train = pd.read_csv('Kaggle_Info/train.csv')
train_backup = train.copy()
test = pd.read_csv('Kaggle_Info/test.csv')
train.head()
```

```python
### Drop columns with >20% null values

null_stats = train.isnull().sum()
more_20_pct_null = null_stats[null_stats > .2 * len(train)]
train.drop(more_20_pct_null.index, axis=1, inplace=True)
more_20_pct_null
```

```python
### Drop columns with uninformative data and leaking info

unwanted_cols = ['Id', 'SaleType', 'SaleCondition', 'MiscVal']
train.drop(unwanted_cols, axis=1, inplace=True)
train.head()
```

```python
### Apply relevant conversions to numerical & categorical cols

cat_cols = ['MSSubClass', 'MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope',
           'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'OverallQual', 'RoofStyle',
           'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
           'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC',
           'CentralAir', 'Electrical', 'KitchenQual', 'Functional', 'GarageType', 'GarageFinish', 'GarageQual',
           'GarageCond', 'PavedDrive']
num_cols = ['LotFrontage', 'LotArea', 'OverallQual', 'OverallCond', 'YearRemodAdd', 'YearBuilt', 'MasVnrArea',
           'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
           'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr',
           'TotRmsAbvGrd', 'Fireplaces', 'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF',
           'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MoSold', 'YrSold']

train[cat_cols] = train[cat_cols].astype('category')
train[num_cols] = train[num_cols].apply(pd.to_numeric)

train.head()
```

```python
### Find columns with low uniqueness (single unique values > 80%)

low_uniqueness_cols = []
for col in train.columns:
    value_counts = train[col].value_counts()
    low_uniqueness_idx = value_counts.max() / value_counts.sum()
    if low_uniqueness_idx >= .8:
        low_uniqueness_cols.append(col)
low_uniqueness_cols
```

```python
### Let's take a look at numerical cols with low uniqueness correlation

num_lu_cols = list(set(low_uniqueness_cols) & set(num_cols))
corr = train[num_lu_cols + ['SalePrice']].corr()
sns.heatmap(corr, vmin=-1, vmax=1)
```

```python
### All of them are correlated poorly, so remove them from features

train.drop(num_lu_cols, axis=1, inplace=True)
```

```python
### Create dummies and inspect categorical cols

cat_lu_cols = list(set(low_uniqueness_cols) & set(cat_cols))
cat_lu_train = pd.get_dummies(train[cat_lu_cols])
cat_lu_train['SalePrice'] = train['SalePrice']
cat_lu_train.corr()['SalePrice'].abs().sort_values(ascending=False).head(10)
```

```python
### ['GarageCond_TA', 'CentralAir_Y', 'GarageQual_TA', 'Electrical_SBrkr', 'PavedDrive_Y'] 
### have pretty good correlation with 'SalePrice'. Let's inspect their heatmap

best_corr_lu_cols = ['GarageCond_TA', 'CentralAir_Y', 'GarageQual_TA', 'Electrical_SBrkr', 'PavedDrive_Y'] 
corr = cat_lu_train[best_corr_lu_cols + ['SalePrice']].corr()
sns.heatmap(corr, vmin=0, vmax=1)
plt.show()
corr
```

*GarageCond_TA* is the most promising column. Considering that it's also strongly correlated with *GarageQual_TA* & *PavedDrive_Y*, let's remove them from features. Let's also keep CentralAir_Y & Electrical_SBrkr for now.

```python
train.drop(cat_lu_cols, axis=1, inplace=True)
cat_lu_train.drop('SalePrice', axis=1, inplace=True)
train = pd.concat([train, cat_lu_train[['GarageCond_TA', 'CentralAir_Y', 'Electrical_SBrkr']]], axis=1)
train.head()
```

Let's now finish dealing with null values. At first let's look at how columns with nulls correlate with *SalePrice*

```python
null_stats = train.isnull().sum().sort_values(ascending=False)
cols_with_nulls = null_stats[null_stats > 0] / len(train)
cols_with_nulls
```

```python
### Start with numeric cols

num_null_cols = list(set(cols_with_nulls.index) & set(num_cols))
corr = train[num_null_cols + ['SalePrice']].corr()
sns.heatmap(corr)
corr
```

*LotFrontage* has pretty strong correlation with our target column. Let's see if we can substitute missing values with mean

```python
train['LotFrontage'].describe()#.describe(percentiles=[.1, .9])
```

Considering that std is pretty low, let's keep *LotFrontage*. Other numeric columns have relatively low pct of null values, so let's keep them too.

```python
train[num_null_cols] = train[num_null_cols].fillna(train[num_null_cols].mean())
train[num_null_cols].info()
```

Let's also keep all null categorical columns (pct of null vals < 5%). Fill null values with mode

```python
cat_null_cols = list(set(cols_with_nulls.index) & set(cat_cols))
train[cat_null_cols] = train[cat_null_cols].fillna(train[cat_null_cols].mode().iloc[0])
train[cat_null_cols].info()
```

```python
train.head()
```

Year columns are not informative by themselves, so let's turn them into intervals

```python
train['YearsBeforeRemodAdd'] = train['YearRemodAdd'] - train['YearBuilt']
train['YearsBeforeGarageBlt'] = train['GarageYrBlt'] - train['YearBuilt']
train['YearsBeforeSold'] = train['YrSold'] - train['YearRemodAdd']
train.drop(['YearBuilt', 'YearRemodAdd', 'YrSold', 'MoSold', 'GarageYrBlt'], axis=1, inplace=True)
sale_price = train.pop('SalePrice')
train['SalePrice'] = sale_price
train.head()
```

Let's inspect numerical cols with abs correlation coefs > .5

```python
num_cols_train = train.select_dtypes(include=[np.number])
strong_cor_cols = num_cols_train.corr()['SalePrice'].abs().sort_values(ascending=False)
strong_cor_cols = strong_cor_cols[strong_cor_cols > .5]
corr = train[strong_cor_cols.index].corr()
sns.heatmap(corr, center=0, vmin=-1, vmax=1)
strong_cor_cols
```

As can be seen on hetmap, there is strong correlation between *GarageArea* and *GarageCars* and also between *GrLiveArea* and *TotRmsAbvGrd*. Let's remove *TotRmsAbvGrd* and *GarageArea* as weaker correlates and also all num cols with corr < .5

```python
strong_corr_num_train = train[strong_cor_cols.index].copy()
strong_corr_num_train.drop(['TotRmsAbvGrd', 'GarageArea'], axis=1, inplace=True)
train.drop(num_cols_train.columns, axis=1, inplace=True)
train = pd.concat([train, strong_corr_num_train], axis=1)
train.head()
```

Let's now inspect categorical cols with abs correlation coefs > .5

```python
cat_cols_train = train.select_dtypes(include='category')
cat_cols_dummies = pd.get_dummies(cat_cols_train)
cat_cols_dummies['SalePrice'] = train['SalePrice']
strong_cor_cols = cat_cols_dummies.corr()['SalePrice'].abs().sort_values(ascending=False)
strong_cor_cols = strong_cor_cols[strong_cor_cols > .5]
corr = cat_cols_dummies[strong_cor_cols.index].corr()
sns.heatmap(corr, center=0, vmin=-1, vmax=1)
corr
```

Leaving only these cat cols

```python
train.drop(cat_cols_train, axis=1, inplace=True)
train.drop('SalePrice', axis=1, inplace=True)
train = pd.concat([train, cat_cols_dummies[strong_cor_cols.index]], axis=1)
sale_price = train.pop('SalePrice')
train['SalePrice'] = sale_price
train.head()
```

Let's combine all our cleaning and transformation in one function and apply it to train and test sets

```python
def clean_and_transform(train, test):
    dfs = []
    
    cat_cols = ['ExterQual', 'BsmtQual', 'KitchenQual', 'GarageFinish']
    num_cols = ['OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', '1stFlrSF', 'FullBath']
    means = train[num_cols].mean()
    modes = train[cat_cols].mode().iloc[0]
    num_cols.append('YearsBeforeSold')

    for orig_df in [train, test]:
        df = orig_df.copy()
        df['YearsBeforeSold'] = df['YrSold'] - df['YearRemodAdd']
        df = df[cat_cols + num_cols].copy()
        df[num_cols] = df[num_cols].fillna(means, axis=0)
        df[cat_cols] = df[cat_cols].fillna(modes, axis=0)
        df[cat_cols] = df[cat_cols].astype('category')
        cat_dummies = pd.get_dummies(df[cat_cols])
        df.drop(cat_cols, axis=1, inplace=True)
        df = pd.concat([df, cat_dummies], axis=1)
        dummy_cat_cols = ['ExterQual_TA', 'BsmtQual_Ex', 'KitchenQual_TA', 'GarageFinish_Unf', 'KitchenQual_Ex']
        df = df[num_cols + dummy_cat_cols].copy()
        dfs.append(df)
    dfs[0] = dfs[0].join(train['SalePrice'])
    return dfs
    
t_train, t_test = clean_and_transform(train_backup, test)
t_test.head()
```

Let's now save cleaned and tranformed features on disk

```python
t_train.to_csv('cleaned_transformed/train.csv')
t_train.to_csv('cleaned_transformed/test.csv')
```
