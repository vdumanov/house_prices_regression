```python
def null_cnt(col):
    return col.isnull().sum()

def null_pct(col):
    return null_cnt(col) / len(col)
```

```python
train = pd.read_csv('../Kaggle_Info/train.csv')
research = train.copy()
test = pd.read_csv('../Kaggle_Info/test.csv')
research.head()
```

```python
def fillna(train, test):
    fill_na_cols = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu',
                    'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'MasVnrType']
    fill_zero_cols = ['LotFrontage', 'LotFrontage', 'GarageYrBlt', 'MasVnrArea']
    train_medians = train.median().iloc[0]
    for df in (train, test):
        unwanted_cols = ['SaleType', 'SaleCondition']
        research.drop(unwanted_cols, axis=1, inplace=True)
        df[fill_na_cols] = df[fill_na_cols].fillna('NA')
        df[fill_zero_cols] = df[fill_zero_cols].fillna(0)
        df['Eletrical'] = df['Electrical'].fillna(train_medians['Electrical'])
        
        bsmt_cols = [col for col in df.columns if 'Bsmt' in col]
        bsmt_null_df = df.loc[df['BsmtQual'].isnull(), bsmt_cols]
```
